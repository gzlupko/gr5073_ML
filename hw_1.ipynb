```python
 # libraries used 
 import pandas as pd 
 import numpy as np 
 import matplotlib as mpl
 import seaborn as sns 
 import matplotlib.pyplot as plt
```

### 1. Find the url for the mtcars dataset fromm the following website: 
https://vincentarelbundock.github.io/Rdatasets/datasets.html 


```python
df = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/mtcars.csv") 
```

### 2. Display the first 5 rows 

```python
df.head(5) 
```

### 3. Calculate the average of the mpg column for all cars within each category of the cyl column. 

```python
# first group by cyl then calculate the average of mpg 

df.groupby('cyl')[['mpg']].mean()

```

### 4. Create a histogram using the mpg column 

```python
df['mpg'].hist()
plt.title('MPG Distribution')
plt.xlabel('MPG')
plt.ylabel('Count')  
```

### 5. Choose two variables in the data set and creat a scatterplot. 

```python
df.head()
sns.scatterplot(data = df, x = "hp", y = "disp") 
plt.title("Horsepower vs. Displacement")
```
